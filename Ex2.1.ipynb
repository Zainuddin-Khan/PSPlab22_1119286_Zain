```python
import pandapower as pp

net = pp.create_empty_network(sn_mva=100)

bus0 = pp.create_bus(net, name="bus0", vn_kv=22, type="n")
bus1 = pp.create_bus(net, name="bus1", vn_kv=22, type="b")
bus2 = pp.create_bus(net, name="bus2", vn_kv=11, type="b")
bus3 = pp.create_bus(net, name="bus3", vn_kv=220, type="b")
bus4 = pp.create_bus(net, name="bus4", vn_kv=220, type="b")
bus5 = pp.create_bus(net, name="bus5", vn_kv=110, type="b")
bus6 = pp.create_bus(net, name="bus6", vn_kv=110, type="b")

gen = pp.create_ext_grid(net, bus0, vm_pu=1, name="Gen")

loads = pp.create_load_from_cosphi(net, bus2, sn_mva=57, const_z_percent=100, cos_phi=0.6, mode="underexcited", name="3ph load", scaling=1.108)

Xs = pp.create_impedance(net, from_bus=0, to_bus=1, sn_mva=90, rft_pu=0, xft_pu=0.18, name="Xs")

T1 = pp.create_transformer_from_parameters(net, hv_bus=3, lv_bus=1, sn_mva=50, vn_hv_kv=220, vn_lv_kv=22,vkr_percent=0,vk_percent=10,pfe_kw=0,i0_percent=0, name="T1")
T2 = pp.create_transformer_from_parameters(net, hv_bus=4, lv_bus=2, sn_mva=40, vn_hv_kv=220, vn_lv_kv=11,vkr_percent=0,vk_percent=6,pfe_kw=0,i0_percent=0, name="T2")
T3 = pp.create_transformer_from_parameters(net, hv_bus=5, lv_bus=1, sn_mva=40, vn_hv_kv=110, vn_lv_kv=22,vkr_percent=0,vk_percent=6.4,pfe_kw=0,i0_percent=0, name="T3")
T4 = pp.create_transformer_from_parameters(net, hv_bus=6, lv_bus=2, sn_mva=40, vn_hv_kv=110, vn_lv_kv=11,vkr_percent=0,vk_percent=8,pfe_kw=0,i0_percent=0, name="T4")

l1 = pp.create_line_from_parameters(net, bus3, bus4, length_km=1,r_ohm_per_km=0,x_ohm_per_km=48.4, c_nf_per_km=0, max_i_ka=0.454, name="Line1")
l2 = pp.create_line_from_parameters(net, bus5, bus6, length_km=1,r_ohm_per_km=0,x_ohm_per_km=65.43, c_nf_per_km=0, max_i_ka=0.909,name="Line2")

pp.runpp(net,algorithm="bfsw",max_iteration=100)
```
